```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(10, 6))

# Рисуем сервер
server = patches.Rectangle((1, 3), 2, 2, linewidth=2, edgecolor='blue', facecolor='lightblue', label='Сервер с антивирусом')
ax.add_patch(server)
plt.text(1.5, 4, 'Сервер\n(Антивирус)', ha='center', va='center', fontsize=10, color='black')

# Рисуем компьютеры с доступом в интернет
internet_pcs = [
    patches.Rectangle((4, 4), 1, 1, linewidth=2, edgecolor='green', facecolor='lightgreen'),
    patches.Rectangle((6, 4), 1, 1, linewidth=2, edgecolor='green', facecolor='lightgreen')
]
for pc in internet_pcs:
    ax.add_patch(pc)
plt.text(4.5, 4.5, 'PC 1\n(Интернет)', ha='center', va='center', fontsize=9, color='black')
plt.text(6.5, 4.5, 'PC 2\n(Интернет)', ha='center', va='center', fontsize=9, color='black')

# Рисуем компьютеры без доступа в интернет
no_internet_pcs = [
    patches.Rectangle((4, 2), 1, 1, linewidth=2, edgecolor='red', facecolor='lightcoral'),
    patches.Rectangle((6, 2), 1, 1, linewidth=2, edgecolor='red', facecolor='lightcoral')
]
for pc in no_internet_pcs:
    ax.add_patch(pc)
plt.text(4.5, 2.5, 'PC 3\n(Нет интернета)', ha='center', va='center', fontsize=9, color='black')
plt.text(6.5, 2.5, 'PC 4\n(Нет интернета)', ha='center', va='center', fontsize=9, color='black')

# Рисуем общий диск
shared_disk = patches.Circle((2, 1), 0.5, edgecolor='purple', facecolor='violet', label='Общий диск')
ax.add_patch(shared_disk)
plt.text(2, 1, 'Общий\nдиск', ha='center', va='center', fontsize=10, color='black')

# Рисуем почтовый сервер
mail_server = patches.Rectangle((0, 1), 1, 1, linewidth=2, edgecolor='orange', facecolor='yellow', label='Почтовый сервер')
ax.add_patch(mail_server)
plt.text(0.5, 1.5, 'Почтовый\nсервер', ha='center', va='center', fontsize=9, color='black')

# Рисуем стрелки (пути распространения вирусов)
plt.arrow(0.5, 2, 1.5, 1, head_width=0.2, head_length=0.2, fc='red', ec='red', label='Распространение вирусов')
plt.arrow(2.5, 3, 1.5, 1, head_width=0.2, head_length=0.2, fc='red', ec='red')
plt.arrow(2.5, 3, 1.5, -1, head_width=0.2, head_length=0.2, fc='red', ec='red')
plt.arrow(2, 1.5, 0, 1.5, head_width=0.2, head_length=0.2, fc='red', ec='red')

# Настройка графика
ax.set_xlim(0, 8)
ax.set_ylim(0, 6)
ax.set_aspect('equal')
ax.axis('off')  # Скрываем оси

# Легенда
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.3, 1))

# Заголовок
plt.title('Схема защиты компьютерной сети', fontsize=14, pad=20)

# Показать график
plt.show()
```
