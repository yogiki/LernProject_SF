```python
import pandas as pd

# Читаем данные
df = pd.read_csv('data/solarpower.csv')

# Меняем -1 на NaN для удобства работы
df['cum_power'] = df['cum_power'].replace(-1, pd.NA)

# Преобразуем в числовой тип, если не был таковым
df['cum_power'] = pd.to_numeric(df['cum_power'], errors='coerce')

# Заполняем пропуски средним значением между соседями
df['cum_power'] = df['cum_power'].interpolate(method='linear')

# Вычисляем среднее и округляем
mean_cum_power = round(df['cum_power'].mean())

# Вывод результата
print(mean_cum_power)
```

```python

# Парсим даты и преобразуем в формат datetime
df['date'] = pd.to_datetime(df['date'])

# Сортируем по дате, если не отсортировано
df = df.sort_values('date').reset_index(drop=True)

# Вычисляем ежедневную выработку как разницу cum_power между днями
df['cum_power'] = df['cum_power'].astype(float)
df['day_power'] = df['cum_power'].diff()

# Убираем NaN, который образуется на первой строке
df = df.dropna(subset=['day_power'])

# Добавляем столбец с периодом год-месяц
df['month_year'] = df['date'].dt.to_period('M')

# Группируем по месяцам и считаем среднее значение day_power
monthly_avg = df.groupby('month_year')['day_power'].mean().round(0).astype(int)

# Выводим результат для марта 2017 года
march_2017_avg = monthly_avg[pd.Period('2017-03', freq='M')]
print(march_2017_avg)
```
