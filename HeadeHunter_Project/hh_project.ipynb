Начнём с простого — с признака «Образование и ВУЗ». Его текущий формат — это: <Уровень образования год выпуска ВУЗ специальность...>. Например:

Высшее образование 2016 Московский авиационный институт (Национальный исследовательский университет)
Неоконченное высшее образование 2000 Балтийская государственная академия рыбопромыслового флота
Нас будет интересовать только уровень образования. Причём, для простоты мы будем считать, что соискатель указывает свой максимальный уровень образования в первых трёх словах столбца «Образование и ВУЗ».

Создайте с помощью функции-преобразования новый признак «Образование», который должен иметь четыре категории: «высшее», «неоконченное высшее», «среднее специальное» и «среднее».

Выполните преобразование, ответьте на контрольные вопросы и удалите признак «Образование и ВУЗ».

```python
# Создаём копию данных
hh_data_copy = hh_data.copy()

# Функция для извлечения уровня образования
def extract_education_level(text):
    # Разделяем текст на слова и берём первые три
    words = text.split()[:3]
    # Объединяем их в строку
    education_level = ' '.join(words)
    
    # Определяем категорию
    if 'высшее' in education_level.lower() and 'неоконченное' not in education_level.lower():
        return 'высшее'
    elif 'неоконченное высшее' in education_level.lower():
        return 'неоконченное высшее'
    elif 'среднее специальное' in education_level.lower():
        return 'среднее специальное'
    elif 'среднее' in education_level.lower():
        return 'среднее'
    else:
        return 'не указано'

# Проверяем пропущенные значения в столбце «Образование и ВУЗ»
print("Количество пропущенных значений:", hh_data_copy['Образование и ВУЗ'].isnull().sum())

# Заменяем пропущенные значения на пустую строку
hh_data_copy['Образование и ВУЗ'] = hh_data_copy['Образование и ВУЗ'].fillna('')

# Применяем функцию к столбцу «Образование и ВУЗ»
hh_data_copy['Образование'] = hh_data_copy['Образование и ВУЗ'].apply(extract_education_level)

# Проверяем уникальные значения нового столбца
print("Уникальные значения в столбце «Образование»:", hh_data_copy['Образование'].unique())

# Удаляем исходный столбец «Образование и ВУЗ»
hh_data_copy.drop(columns=['Образование и ВУЗ'], inplace=True)

# Проверяем результат
hh_data_copy.head()
```

Следующим этапом преобразуем признак «Опыт работы». Его текущий формат: <Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>.

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовём «Опыт работы (месяц)».

Для начала обсудим условия решения задачи:

Во-первых, в данном признаке есть пропуски. Условимся, что, если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN).
Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значение Не указано. Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях).

В-четвёртых, у нас есть проблема: опыт работы может быть представлен только в годах или только в месяцах. Например, можно встретить следующие варианты:

Опыт работы 3 года 2 месяца
Опыт работы 4 года
Опыт работы 11 месяцев
Учитывайте эту особенность в вашем коде.

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.

Выполните преобразование и удалите столбец «Опыт работы» из таблицы.

```python
import numpy as np

# Функция для преобразования опыта работы в месяцы
def get_experience(arg):
    # Если значение пропущено или равно "Не указано", возвращаем NaN
    if pd.isna(arg) or arg == 'Не указано':
        return np.nan
    
    # Ключевые слова для поиска лет и месяцев
    year_words = ['год', 'года', 'лет']
    month_words = ['месяц', 'месяца', 'месяцев']
    
    # Разделяем строку на части (первые 7 элементов)
    arg_splitted = arg.split(' ')[:7]
    years = 0
    months = 0
    
    # Проходим по частям строки
    for index, item in enumerate(arg_splitted):
        # Если находим ключевое слово из years_Word извлекаем количество лет
        if item in year_words:
            try:
                years = int(arg_splitted[index - 1]) 
            except (ValueError, IndexError):
                years = 0  # Если не удалось преобразовать, считаем 0
        
        # Если находим ключевое слово из month_words извлекаем количество месяцев
        if item in month_words:
            try:
                months = int(arg_splitted[index - 1])  # Берём предыдущий элемент (число)
            except (ValueError, IndexError):
                months = 0  # Если не удалось преобразовать, считаем 0
    
    # Преобразуем годы в месяцы и суммируем
    return int(years * 12 + months)

# Применяем функцию к столбцу «Опыт работы»
hh_data_copy['Опыт работы (месяц)'] = hh_data_copy['Опыт работы'].apply(get_experience)

# Выводим два столбца для проверки
hh_data_copy[['Опыт работы', 'Опыт работы (месяц)']].head(10)  # Первые 10 строк

# Удаляем исходный столбец «Опыт работы» (если нужно)
# hh_data_copy.drop(columns=['Опыт работы'], inplace=True)

```

Рассмотрим поближе признаки «Занятость» и «График».

Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удалённая работа, вахтовый метод).

На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:

полная занятость, частичная занятость;
частичная занятость, проектная работа, волонтерство;
полный день, удалённая работа;
вахтовый метод, гибкий график, удалённая работа, полная занятость.
Такой вариант признаков имеет множество различных комбинаций, а значит, множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-«мигалки» для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе — False.

Такой метод преобразования категориальных признаков называется One Hot Encoding, его схема представлена на рисунке ниже:

Выполните данное преобразование для признаков «Занятость» и «График», ответьте на контрольные вопросы, после чего удалите их из таблицы.

```python
# Список уникальных категорий для «Занятость» и «График»
employment_categories = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule_categories = ['полный день', 'сменный график', 'гибкий график', 'удалённая работа', 'вахтовый метод']

# Функция для создания признаков-«мигалок»
def create_ohe_columns(df, column, categories):
    for category in categories:
        df[f'{column}_{category}'] = df[column].apply(lambda x: category in x)
    return df

# Применяем One Hot Encoding к столбцу «Занятость»
hh_data_copy = create_ohe_columns(hh_data_copy, 'Занятость', employment_categories)

# Применяем One Hot Encoding к столбцу «График»
hh_data_copy = create_ohe_columns(hh_data_copy, 'График', schedule_categories)

# Удаляем исходные столбцы «Занятость» и «График»
hh_data_copy.drop(columns=['Занятость', 'График'], inplace=True)

# Проверяем результат
hh_data_copy.head()
```

Умножить сумму желаемой заработной платы на присоединённый курс валюты (close) и разделить на пропорцию. Обратите внимание на пропуски после объединения в этих столбцах. Результат занести в новый столбец «ЗП (руб)».

Удалите исходный столбец заработной платы «ЗП» и все промежуточные столбцы, если вы их создавали.

```python
import pandas as pd

# Преобразуем столбец «Обновление резюме» в datetime
hh_data_copy['Обновление резюме'] = pd.to_datetime(hh_data_copy['Обновление резюме'], format='%d.%m.%Y %H:%M')

# Извлекаем только дату
hh_data_copy['Дата обновления'] = hh_data_copy['Обновление резюме'].dt.date

# Загружаем данные о курсах валют
currency_data = pd.read_csv('data/currency_rates.csv')

# Преобразуем столбец date в datetime с правильным форматом
currency_data['date'] = pd.to_datetime(currency_data['date'], format='%d/%m/%y').dt.date

# Оставляем только нужные столбцы
currency_data = currency_data[['currency', 'date', 'proportion', 'close']]

# Создаем словарь соответствия валют по заданию
currency_mapping = {
    'грн': {'iso': 'UAH', 'proportion': 10},
    'USD': {'iso': 'USD', 'proportion': 1},
    'EUR': {'iso': 'EUR', 'proportion': 1},
    'белруб': {'iso': 'BYN', 'proportion': 30},
    'KGS': {'iso': 'KGS', 'proportion': 10},
    'сум': {'iso': 'UZS', 'proportion': 10000},
    'AZN': {'iso': 'AZN', 'proportion': 1},
    'KZT': {'iso': 'KZT', 'proportion': 100},
    'руб': {'iso': 'RUB', 'proportion': 1}  # Добавляем рубль
}

# Разделяем столбец «ЗП» на сумму и валюту
hh_data_copy[['ЗП сумма', 'ЗП валюта']] = hh_data_copy['ЗП'].str.split(n=1, expand=True)

# Удаляем точку из валюты
hh_data_copy['ЗП валюта'] = hh_data_copy['ЗП валюта'].str.replace('.', '')

# Преобразуем в float
hh_data_copy['ЗП сумма'] = hh_data_copy['ЗП сумма'].str.replace(' ', '').astype(float)

# Используем словарь currency_mapping для преобразования названий валют в ISO-коды
hh_data_copy['ЗП валюта'] = hh_data_copy['ЗП валюта'].map(
    lambda x: currency_mapping[x]['iso'] if x in currency_mapping else None
)

# Проверяем результат
print("Уникальные значения в 'ЗП валюта' после преобразования:")
print(hh_data_copy['ЗП валюта'].unique())

# Объединяем данные с курсами валют
merged_data = pd.merge(
    hh_data_copy,
    currency_data,
    left_on=['Дата обновления', 'ЗП валюта'],
    right_on=['date', 'currency'],
    how='left'
)

# Заполняем proportion и close для рубля (RUB)
merged_data.loc[merged_data['ЗП валюта'] == 'RUB', 'close'] = 1
merged_data.loc[merged_data['ЗП валюта'] == 'RUB', 'proportion'] = 1.0

# Заполняем NaN в proportion и close значением 1 (для валют, которые не нашлись в currency_data)
merged_data['close'] = merged_data['close'].fillna(1)
merged_data['proportion'] = merged_data['proportion'].fillna(1)

# Рассчитываем ЗП в рублях
merged_data['ЗП (руб)'] = (merged_data['ЗП сумма'] * merged_data['close']) / merged_data['proportion']

# Удаляем лишние столбцы
merged_data.drop(
    columns=['ЗП', 'ЗП сумма', 'ЗП валюта', 'currency', 'date', 'proportion', 'close'],
    inplace=True
)

# Проверяем медианную зарплату
median_salary = round(merged_data['ЗП (руб)'].median() / 1000)
print(f"Медианная заработная плата: {median_salary} тысяч рублей")
```

Постройте распределение признака «Возраст». Опишите распределение, отвечая на следующие вопросы:

Чему равна мода распределения?
Каковы предельные значения признака, в каком примерном интервале находится возраст большинства соискателей?
Есть ли аномалии для данного признака? Если есть, то какие значения вы бы причислили к таковым?

Постройте гистограмму и коробчатую диаграмму рядом.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Преобразование возраста в числовой формат
merged_data['Возраст_число'] = merged_data['Возраст'].str.split().str[0].astype(int)

# Настройка стиля графиков через seaborn
sns.set_style("whitegrid")  # Устанавливаем стиль seaborn
plt.figure(figsize=(12, 6))

# Гистограмма распределения
plt.subplot(1, 2, 1)
sns.histplot(data=merged_data, x='Возраст_число', bins=20, kde=True)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество соискателей')

# Box plot
plt.subplot(1, 2, 2)
sns.boxplot(data=merged_data, x='Возраст_число')
plt.title('Коробчатая диаграмма возраста')
plt.xlabel('Возраст')

plt.tight_layout()
plt.show()

# Статистические характеристики
mode_age = merged_data['Возраст_число'].mode().values[0]
min_age = merged_data['Возраст_число'].min()
max_age = merged_data['Возраст_число'].max()
q25, q75 = merged_data['Возраст_число'].quantile([0.25, 0.75])

print(f"Мода распределения: {mode_age} лет")
print(f"Минимальный возраст: {min_age} лет")
print(f"Максимальный возраст: {max_age} лет")
print(f"25-й перцентиль: {q25:.0f} лет")
print(f"75-й перцентиль: {q75:.0f} лет")
```
